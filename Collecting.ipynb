```python
from urllib.request import urlopen
GitLinkSAV="https://github.com/reconjohn/evans/raw/master/hsb_ok.sav"
response = urlopen(GitLinkSAV) # open link
content = response.read() # read contents
fhandle = open('savFromGit.sav', 'wb') # open a file
fhandle.write(content) # save contents in that file
fhandle.close() # close the file
```
