```python
import pandas as pd

GitLinkXLS='https://github.com/reconjohn/evan_data/raw/master/energy-use-per-capita.csv'

EnergyConsume=pd.read_csv(GitLinkXLS)[pd.read_csv(GitLinkXLS).Year==2014]
EnergyConsume.drop(columns='Year',inplace=True)
```

```python
EnergyConsume
```

```python
GitLinkXLS='https://github.com/reconjohn/evan_data/raw/master/installed-solar-PV-capacity.csv'

pd.read_csv(GitLinkXLS).head()
SolarCapa=pd.read_csv(GitLinkXLS)[pd.read_csv(GitLinkXLS).Year==2015]
SolarCapa.drop(columns='Year',inplace=True)
SolarCapa
```

```python
GitLinkXLS='https://github.com/reconjohn/evan_data/raw/master/renewable-energy-consumption.csv'

SolarConsume=pd.read_csv(GitLinkXLS)[pd.read_csv(GitLinkXLS).Year==2015].loc[:,['Entity','Code','Year','Solar (terawatt-hours)']]
SolarConsume.drop(columns='Year',inplace=True)
SolarConsume.head()
```

```python
len(EnergyConsume), len(SolarCapa), len(SolarConsume)
```

```python
Solar=SolarCapa.merge(EnergyConsume,how='left').merge(SolarConsume,how='left')
```

```python
Solar.columns
```

```python
Solar.columns=Solar.columns.str.replace("\s+|\(|\)","").str.replace("gigawatts","GW").str.replace("Entity","Country").str.replace("terawatt-hours","TWh")
```

```python
OldToNew={'EnergyusekgofoilequivalentpercapitakWh':'EnergyUseKWh','SolarTWh':'SolarUseTWh'}
```

```python
Solar.rename(columns=OldToNew,inplace=True)
Solar
```

```python
Solar.to_csv('SolarData.csv', index=False)
```
