```python
import pandas as pd
wikiLink="https://en.wikipedia.org/wiki/List_of_countries_by_electricity_production_from_renewable_sources" 
DFs=pd.read_html(wikiLink,header=0,flavor='bs4',attrs={'class': 'wikitable'})
```

```python
DF=DFs[0].copy()
```

```python
DF.columns
```

```python
DF.iloc[:, [0,3,14,16]].columns
```

```python
pattern="\\s+|\(|\)"
replacer=""

DF.iloc[:, [0,3,14,16]].columns.str.replace(pattern,replacer).str.replace("Solarpower","SolarGWh").str.replace("SolarGWh.2","SolarPct")
```

```python
NewNames=DF.iloc[:, [0,3,14,16]].columns.str.replace(pattern,replacer).str.replace("Solarpower","SolarGWh").str.replace("SolarGWh.2","SolarPct")
list(zip(DF.iloc[:, [0,3,14,16]].columns,NewNames))
```

```python
changer={old:new for old,new in zip(DF.iloc[:, [0,3,14,16]].columns,NewNames)}
changer
```

```python
DF.rename(columns=changer,inplace=True)
```

```python
data=DF.iloc[:, [0,3,14,16]].drop(index=0)
```

```python
result1=data.SolarPct.str.split('%',expand=True)[0]
```

```python
data=data.assign(SolarPct=result1)
```

```python
data
```

```python
badSymbols=[]
for column in data.iloc[:,1::]:
    for cell in data[column]:
        try:
            float(cell)
        except:
            badSymbols.append(cell)
```

```python
badSymbols
```

## Merging education data

```python
GitLinkXLS='https://raw.githubusercontent.com/cdquack/543/master/2015EI2.csv'

Education=pd.read_csv(GitLinkXLS)
Education
```

```python
Education.Country.values
```

```python
data.Country.values
```

```python
# for energy data to covert
country_fix = {'Czechia': 'Czech Republic',
               'Korea Rep': 'South Korea',
               'Russia': 'Russian Federation',
               'Brunei': 'Brunei Darussalam',
                'North Macedonia': 'Macedonia',
               'Vietnam': 'Viet Nam',
               'Eswatini (Swaziland)': 'Swaziland',
                'St Lucia': 'Saint Lucia',
               'Guinea Bissau': 'Guinea-Bissau',
               'Republic of the Congo': 'Congo',
               'Syria': 'Syrian Arab Republic',
                'Sao Tome & Principe': 'Sao Tome and Principe',
               'St Vincent & the Grenadines':'Saint Vincent and the Grenadines',
               'St Kitts and Nevis':'Saint Kitts and Nevis',
                'Timor Leste':'Timor-Leste'}

# use the dictionary
data.Country.replace(country_fix,inplace=True)
```

```python
merged=data.merge(Education, how='outer',indicator=True)
merged.loc[merged['_merge']=='right_only',"Country"]
```

```python
merged.loc[merged['_merge']=='left_only',"Country"]
```

```python
merged_data=merged.loc[merged['_merge']=='both',]
merged_data.drop(columns='_merge',inplace=True)
```

```python
merged_data
```

```python
Data=merged_data.dropna()
Data.reset_index(drop=True,inplace=True)
Data
```
